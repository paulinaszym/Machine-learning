Zadanie 7<br>
Ze względu na to, że będziemy wykonywać to samo dla liczby zachorowań i liczby śmierci tworzenie modeli zrobiłam w funkcji 'lin_reg', która zwraca otrzymane wyniki dla wykonanych modeli regresji liniowej.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import pandas as pd
import numpy as np

def lin_reg(col,df,cols1,cols2):
  """Function that creates a linear regression 
    models for all given columns and returns 
    a dataframe with correlation value and
    linear regression metrics values. """
    
  res = pd.DataFrame(columns=cols2)
  y=df[col]
  df_corr=df.corr()
  for c2 in cols1:         
    x=df[[c2]]
    X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
    LR=LinearRegression()
    LR.fit(X_train,y_train)
    PredTest=LR.predict(X_test)
    s=LR.score(X_test,y_test)
    CORR=df_corr[col][c2]
    VAR=metrics.explained_variance_score(y_test, PredTest)
    ME=metrics.max_error(y_test, PredTest)
    MAE=metrics.mean_absolute_error(y_test,PredTest)
    MSE=metrics.mean_squared_error(y_test,PredTest)
    RMSE=np.sqrt(MSE)
    res=res.append(pd.Series({'Correlation':CORR,'Score':s,'Variance':VAR,'ME':ME,'MAE':MAE,'MSE':MSE,'RMSE':RMSE},name=c2))
  return(res)
```

Aby nie kopiować za każdym razem tego samego kodu zdefiniuję do tego celu funkcję "assumptions_plots".

```python
import matplotlib.pyplot as plt

def assumptions_plots(col1,col2,df_diff):
  """Plots linear regression assumptions"""

  y=df_diff[col1]
  x=df_diff[[col2]]
  X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
  LR=LinearRegression()
  LR.fit(X_train,y_train)
  PredTest=LR.predict(X_test)
  s=LR.score(X_test,y_test)

  fig = plt.figure(figsize=(15, 5))
  ax1 = fig.add_subplot(131)
  ax2 = fig.add_subplot(132)
  ax3 = fig.add_subplot(133)
  axes = [ax1, ax2, ax3]
  
  #Linearity assumption
  axes[0].scatter(x, y, color='k', s=0.7)
  c, i = np.polyfit(x[col2], y, 1)
  axes[0].plot(x, c*x+i)
  axes[0].set_title("Linearity assumption")
  
  #Normal distribution of errors
  residual=y_test-PredTest
  axes[1].hist(residual,edgecolor = 'black')
  axes[1].set_title("Normal distribution of errors")

  #Homoscedasticity
  axes[2].scatter(PredTest,residual)
  axes[2].set_title("Homoscedasticity")
```
